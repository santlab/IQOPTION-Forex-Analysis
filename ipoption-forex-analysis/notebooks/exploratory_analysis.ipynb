# Exploratory Data Analysis for Forex Trading

This notebook is designed for exploratory data analysis (EDA) of Forex trading data. It will help visualize and analyze market trends, indicators, and signals generated from the IPOPTION chart.

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style
sns.set(style='whitegrid')
```

```python
# Load the market data
data = pd.read_csv('path_to_your_forex_data.csv')  # Update with the actual path

# Display the first few rows of the dataset
data.head()
```

```python
# Check for missing values
missing_values = data.isnull().sum()
missing_values[missing_values > 0]
```

```python
# Visualize the price trends
plt.figure(figsize=(14, 7))
plt.plot(data['Date'], data['Close'], label='Close Price')  # Update with actual column names
plt.title('Forex Close Price Trend')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
# Calculate and visualize moving averages
data['SMA_20'] = data['Close'].rolling(window=20).mean()  # Update with actual column names
data['SMA_50'] = data['Close'].rolling(window=50).mean()

plt.figure(figsize=(14, 7))
plt.plot(data['Date'], data['Close'], label='Close Price')
plt.plot(data['Date'], data['SMA_20'], label='20-Day SMA', color='orange')
plt.plot(data['Date'], data['SMA_50'], label='50-Day SMA', color='green')
plt.title('Forex Price with Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```
